**Exercise 1**: The purpose of this exercise is to use neural networks to learn a function y = f(x). You will be given a set of inputs x and outputs y. You will explore different networks as instructed below to see how each of them can capture (or not) the properties of the true underlying function.

```python
# importing neccessasry libraries
import numpy as np
import torch.optim as optim
from matplotlib import pyplot as plt
import torch
import torch.nn as nn
```

```python
# creating data for training

def func(x):
  return x**3 - 0.5 * x**2 - x + 1.5

x = np.linspace(-3, 3, 80)
y = func(x)

x_test = np.concatenate([np.linspace(3.1, 4, 10), np.linspace(-2.5, 2.5, 17)])
y_test = func(x_test)
x, x_test
```

```python
#@title Visualizing the ground truth training data
plt.figure()
plt.scatter(x, y)
plt.show()
```

```python
#@title TODO: Define the neural networks

class NeuralNet_1Layer(nn.Module):
  def __init__(self):
    super(NeuralNet_1Layer, self).__init__()
    self.linear = nn.Linear(in_features=1, out_features=1)

  def forward(self, x):
    out = self.linear(x)
    return out

class NeuralNet_2Layer(nn.Module):
  def __init__(self):
    super(NeuralNet_2Layer, self).__init__()
    # 2 layer neural network, 5 neurons hidden layer
    self.linear1 = nn.Linear(1, 5)
    self.linear2 = nn.Linear(5, 1)

  def forward(self, x):
    layer_in = self.linear1(x)
    out = self.linear2(layer_in)
    return out


class NeuralNet_2Layer_activation(nn.Module):
  def __init__(self):
    super(NeuralNet_2Layer_activation, self).__init__()
    # 2 layer neural network, 5 neurons hidden layer, activation
    self.linear1 = nn.Linear(1, 10)
    self.activation = torch.nn.ReLU()
    self.linear2 = nn.Linear(10, 1)

  def forward(self, x):
    layer_in = self.linear1(x)
    act = self.activation(layer_in)
    out = self.linear2(act)
    return out
```

```python
# convert input and output to torch tensors, so that we can feed them to a PyTorch network
x_tensor = torch.from_numpy(x).reshape(-1, 1).float()
y_tensor = torch.from_numpy(y).reshape(-1, 1).float()
x_test_tensor = torch.from_numpy(x_test).reshape(-1, 1).float()
y_test_tensor = torch.from_numpy(y_test).reshape(-1, 1).float()

# Hyper parameters
epochs = 500
batch_size = 20
learning_rate = 0.01 # a lr of ~0.05 and >10 neurons in the third NN seem to give decent results

# initialize networks and set them to training mode
net1 = NeuralNet_1Layer().train()
net2 = NeuralNet_2Layer().train()
net3 = NeuralNet_2Layer_activation().train()

# initialize optimizers
optimizer1 = torch.optim.Adam(net1.parameters(), learning_rate)
optimizer2 = torch.optim.Adam(net2.parameters(), learning_rate)
optimizer3 = torch.optim.Adam(net3.parameters(), learning_rate)

def training_loop(net, optimizer, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor):
  # initialize loss function
  loss_fn = nn.MSELoss()

  for epoch in range(epochs):
    for i in range(0, len(x_tensor), batch_size):

      optimizer.zero_grad()

      # get input data and ground truth slices for current index
      x_input = x_tensor[i:i+batch_size]
      y_gt = y_tensor[i:i+batch_size]

      # get network prediction for input batch x_input
      y_predicted = net(x_input)

      # compute loss and backpropagate
      loss = loss_fn(y_predicted, y_gt)
      loss.backward()

      # call an optimization step
      optimizer.step()

    # evaluate the network every 10 epochs
    if epoch % 10 == 0:
      y_predicted = net(x_test_tensor)
      error = loss_fn(y_predicted, y_test_tensor).item()
      print("Epoch =", epoch, "Error =", error)

  # setting the network to evaluate mode
  net.eval()

  #compute the mean L2 error on the whole training set
  y_predicted = net(x_tensor)
  error = loss_fn(y_predicted, y_tensor).item()
  print("Final TRAIN error =", error)

  # compute the mean L2 error in the test set
  y_test_predicted = net(x_test_tensor)
  error = loss_fn(y_test_predicted, y_test_tensor).item()
  print("Final TEST error =", error)

  # converting the predictions to numpy so that we can visualize them
  y_pred = y_predicted.reshape(-1).detach().numpy()
  y_test_pred = y_test_predicted.reshape(-1).detach().numpy()

  # vizualize predictions and ground truth data for training and test set
  plt.figure()
  plt.scatter(x, y)
  plt.scatter(x, y_pred)
  plt.scatter(x_test, y_test)
  plt.scatter(x_test, y_test_pred)
  plt.show()

# call the training loop for each of your networks
print("Starting training for 1 layer NN")
training_loop(net1, optimizer1, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor)
print("Starting training for 2 layer NN")
training_loop(net2, optimizer2, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor)
print("Starting training for 2 layer NN with activation")
training_loop(net3, optimizer3, x_tensor, y_tensor, batch_size, epochs, x_test_tensor, y_test_tensor)

```

The only network that is able to capture non-linearities is the third one.

The first network consists of a single linear layer, which won't be able to approximate the cubic function. The second network does have two linear layers, but without the activation function it ends up performing a linear transformation, making the hidden layer redundant. The third one contains an activation function, introducing non-linearities.

Usually, neural networks are great at interpolating (accurate of tests between training samples), because they are trained to approximate based on training samples. Neural networks learn from the patterns present in the test data. The challenge is making them extrapolate well (to make them accurate with tests outside of the training data).
